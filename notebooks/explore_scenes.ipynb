# Scene Browser and Selector

This notebook loads all available driving scenes from `clip_ids.parquet`, visualizes the camera frames, and helps you choose suitable `clip_id` values for inference in `inference.ipynb`.

**Note**: This notebook does not load the model or run inference, so no GPU memory is required.

```python
import pandas as pd
import mediapy as mp

from alpamayo_r1.load_physical_aiavdataset import load_physical_aiavdataset
```

### Browse All Available Scenes

Read `clip_ids.parquet` and display the total number of scenes along with the full list.

```python
df = pd.read_parquet("clip_ids.parquet")
print(f"Total available scenes: {len(df)}")
print(f"Columns: {df.columns.tolist()}")
df
```

### Select and Visualize a Single Scene

Change `selected_idx` to browse different scenes by index, or directly assign a `clip_id` string.

Image layout: 4 images per row, covering 4 cameras (left, front wide, right, front tele), each with 4 frames (from t0-0.3s to t0).

```python
selected_idx = 770
clip_id = df["clip_id"].iloc[selected_idx]

# Alternatively, specify a clip_id directly:
# clip_id = "030c760c-ae38-49aa-9ad8-f5650a545d26"

print(f"Scene index: {selected_idx}, clip_id: {clip_id}")

data = load_physical_aiavdataset(clip_id, t0_us=5_100_000)

mp.show_images(
    data["image_frames"].flatten(0, 1).permute(0, 2, 3, 1),
    columns=4,
    width=200,
)
```

### Batch Preview Multiple Scenes

View camera frames for multiple scenes at once for side-by-side comparison. Modify the `preview_indices` list to select the scenes you want to preview.

```python
preview_indices = [0, 50, 103, 204, 305, 400, 500, 770]

for idx in preview_indices:
    cid = df["clip_id"].iloc[idx]
    print(f"\n--- Scene #{idx}: {cid} ---")
    d = load_physical_aiavdataset(cid, t0_us=5_100_000)
    mp.show_images(
        d["image_frames"].flatten(0, 1).permute(0, 2, 3, 1),
        columns=4,
        width=200,
    )
```
